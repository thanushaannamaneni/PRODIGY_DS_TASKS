# Data Distribution Visualization

## Objective
To visualize the distribution of a categorical and continuous variable using bar chart and histogram.

```
import pandas as pd
import matplotlib.pyplot as plt
```

```
data = {
    "Age": [22, 25, 30, 22, 40, 35, 30, 28, 22, 25],
    "Gender": ["Male", "Female", "Female", "Male", "Male", "Female", "Male", "Female", "Male", "Female"]
}

df = pd.DataFrame(data)
df
```

## Histogram: Age Distribution (Continuous Variable)

```
plt.hist(df['Age'], bins=5)

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

## Bar Chart: Gender Distribution (Categorical Variable)

```
df['Gender'].value_counts().plot(kind='bar')

plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")

plt.show()
```

## Conclusion

- Histogram is used for continuous data like Age.
- Bar chart is used for categorical data like Gender.
- These graphs help understand data distribution.
